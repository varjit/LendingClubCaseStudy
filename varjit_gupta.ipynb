```python
#Printing column with % of missing values to understand the extent of missing values
null_df=(loan_df.isnull().sum()/len(loan_df)*100).sort_values(ascending=False)
null_df.head(48)
```

```python
#Dropping Columns having more than 90% Null Values
loan_df.drop(["next_pymnt_d", "mths_since_last_record"],axis=1,inplace=True)
```

```python
#id - id will not have any effect 
#member_id- id will not have any effect 
#url -  url will not have any effect ,
#desc - has description, which will not have any effect
#last_credit_pull_d - Its last credit pull date, that will not have any effect
#funded_amnt_inv - Investor related, no effect in the outcome
#funded_amnt - Same as loan amount, hence we can drop off

#Dropping above mentioned columns
loan_df.drop(["id", "member_id","url","desc","last_credit_pull_d","funded_amnt_inv","funded_amnt"],axis=1,inplace=True)
```

```python
#emp_title - Doesn't seem to have standard value that can be used for analysis
#zip_code - Zip Codes are masked, hecne should not be used for analysis
#title - This column seem to have non standard comments type value, that may not add value to analysis 
#Dropping Above mentioned columns
loan_df.drop(["emp_title", "zip_code","title"],axis=1,inplace=True)
```

```python
#As we need to find the drivng factors behind the 
#default, hence only loans that are paid off and that are default should be considered
#Hence deleting the row with that has current loans
loan_df = loan_df[loan_df.loan_status != "Current"]
loan_df.nunique()
```

```python
#We can see emp_length,pub_rec_bankruptcies and revol_util are the only column with missing values
#
print("Mode : " + loan_df.emp_length.mode()[0])
loan_df.emp_length.value_counts()
```

```python
print("Mode : " + loan_df.revol_util.mode()[0])
loan_df.revol_util.value_counts()
```

```python
print("Mode : " + str(loan_df.pub_rec_bankruptcies.mode()[0]))
loan_df.pub_rec_bankruptcies.value_counts()
```

```python
#Based on above observations of the mode value and number of rows with missing values less than 3% , we can safely impute the missing values with mode
loan_df.emp_length.fillna(loan_df.emp_length.mode()[0], inplace = True)
loan_df.pub_rec_bankruptcies.fillna(loan_df.pub_rec_bankruptcies.mode()[0], inplace = True)
loan_df.revol_util.fillna(loan_df.revol_util.mode()[0], inplace = True)
```

```python
null_df=(loan_df.isnull().sum()/len(loan_df)*100).sort_values(ascending=False)
null_df.head(30)
```

```python
#Now we don't have any missing values
#Initiating Data Standardization
#Converting revol_util,int_rate to numeric from object type
loan_df.revol_util = pd.to_numeric(loan_df.revol_util.apply(lambda x : x.split('%')[0]))
loan_df.int_rate = pd.to_numeric(loan_df.int_rate.apply(lambda x : x.split('%')[0]))
```

```python
#Converting <1 to 0 and 10+ Years to 10 in emp_length
loan_df.emp_length = pd.to_numeric(loan_df.emp_length.apply(lambda x: 0 if "<" in x else (x.split('+')[0] if "+" in x else x.split()[0])))
```

```python
#Extracting Subgrade
loan_df.sub_grade = pd.to_numeric(loan_df.sub_grade.apply(lambda x : x[1]))
```

# **Bivariate Analysis**

```python
def Bivariate_Categorical (*, data=loan_df, x, y, title=None, figsize=(8,6)):
    """
        Parameters:
            data : defaulted at the 'loan' variable, but can be changed externally if needed
            x, y : columns of data in the corresponding axis, both categorical
            figsize : a default have is given, but can be overridden
        Returns:
            None, does a heatmap plot
    """
    fig, ax = plt.subplots(figsize=figsize)
    pt = pd.pivot_table(data=data, values='loan_status_num', index=y, columns=x)
    sns.heatmap(pt, ax=ax, cmap='YlGnBu')
    if title is not None:
        plt.title(title)
    plt.show()
```

```python
Bivariate_Categorical(x='installment_bin', y='annual_inc_bin')
plt.show()
```

```python
Bivariate_Categorical(x='loan_amnt_bin', y='annual_inc_bin')
plt.show()
```
